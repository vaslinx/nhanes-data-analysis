```python
import pandas as pd 
df=pd.read_csv('nhanes_clean.csv')
table_age = df.groupby('AgeGroup')[['Height', 'Weight']].mean()
print("\nсредний рост и вес по возрастным группам:")
print(table_age)
table_min_max= df[['Height', 'Weight']].agg(['min','max'])
print("\nминимальный и максимальный рост и вес:")
print(table_min_max)
```

```python
import pandas as pd
age_counts=df['AgeGroup'].value_counts()
print("\nколичество людей в каждой возрастной группе:")
print (age_counts)
age_gender_counts=df.groupby(['AgeGroup', 'Gender'])['SEQN'].count()
print("\n количество мужчин и женщин в каждой возрастной группе:")
print(age_gender_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(df['Age'], bins = 10, color='pink', edgecolor='black')
plt.title("распределение возрастов")
plt.xlabel("возраст")
plt.ylabel("количество людей")
plt.show()

sns.countplot(x= 'AgeGroup', hue= 'Gender', data=df, palette='Set2')
plt.title("количество мужчин и женщин по возрастным группам")
plt.xlabel("возрастная группа")
plt.ylabel("количество")
plt.show()
```

```python
import pandas as pd
stats_age=df.groupby('AgeGroup')[['Height', 'Weight']].agg(['mean', 'median', 'std', 'count'])
print("\n statistics for age groups")
print(stats_age)
stats_age_gender = df.groupby(['AgeGroup', 'Gender'])[['Height', 'Weight']].agg(['mean', 'median', 'std', 'count'])
print("\n statistics for age groups and gender")
print(stats_age_gender)
df['BMI']=df['Weight']/(df['Height']/100)**2
print(df[['Age', 'Gender', 'Height','Weight', 'BMI']].head())
```

```python
import pandas as pd 
import seaborn as sns
corr=df[['Age', 'Height', 'Weight', 'BMI']].corr()
print("\n correlation of variables:")
print(corr)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("correlation of variables such as Age, Height, Weight, BMI")
plt.show()
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("quantity of people")
plt.show()
sns.countplot(
    x='AgeGroup',
    data=df,
    color='red'
)

plt.title("Quantity of People by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Quantity")
plt.show()
sns.countplot(
    x='AgeGroup',
    hue='Gender',
    data=df,
    palette='Set2'
)

plt.title("Quantity of Men and Women by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Quantity")
plt.show()
plt.hist(df['BMI'], bins=10, color='pink', edgecolor='black')
plt.title("distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("quantity of people")
plt.show()
sns.boxplot(
    x='AgeGroup',
    y='BMI',
    data=df,
    color='lightgreen'
)

plt.title("BMI by Age Group")
plt.xlabel("Age Group")
plt.ylabel("BMI")
plt.show()
sns.boxplot(
    x='Gender',
    y='BMI',
    data=df,
    color='blue'
)

plt.title("BMI by Gender")
plt.xlabel("Gender")
plt.ylabel("BMI")
plt.show()
df['z_BMI']=(df['BMI']-df['BMI'].mean())/df['BMI'].std()
df_no_outliers=df[df['z_BMI'].abs()<3]
sns.boxplot(
    x='AgeGroup',
    y='BMI',
    hue='AgeGroup',
    data=df_no_outliers,
    palette='pastel',
    dodge=False
)
plt.title("BMI by age group without outliers (z-score)")
plt.xlabel("Age group")
plt.ylabel("BMI")
plt.show()
```
